```python
import cv2
from glob import glob
import matplotlib.pyplot as plt
import json
```

```python
image_dir  = '../road_pranto_bhai'
height = []
width = []
is_landscape = []
for image_name in glob(image_dir + "/*/*/*"):
    try:
        image = cv2.imread(image_name)
        h, w, _ = image.shape
        height.append(h)
        width.append(w)
        if h>w:
            is_landscape.append(1)
        elif h==w:
            is_landscape.append(0)
        else:
            is_landscape.append(-1)
    except:
        pass
```

```python
print(len(glob(image_dir + '/*/train/*')))
print(len(glob(image_dir + '/*/val/*')))
```

```python
print({len(glob(image_dir + "/*/*/*"))})
for val in set(is_landscape):
    print(f"{val}: {is_landscape.count(val)}")
new_h_w = height + width
plt.hist(new_h_w, 5)
plt.xlabel("image height-width")
plt.ylabel("counts in each category")
plt.title("image height-width histogram")
print(min(new_h_w), max(new_h_w))
fig1, ax1 = plt.subplots()
ax1.pie([669, 619, 4], explode = (0, 0, 0.5), labels = ['Landscape', 'Potrait', 'Square'], 
        autopct='%1.1f%%', shadow=True, startangle=90)
```
